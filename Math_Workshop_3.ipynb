```python
# Question 1
import matplotlib.pyplot as plt
import math
x = []
y = []
start = -4 * math.pi
end = 4 * math.pi
step = 0.1
i = start
while i <= end:
    x.append(i)
    y.append(math.sin(i))
    i += step
plt.plot(x, y, color="red", label="sin function")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y = sin(x) ,-4π<= x <= 4π")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Question 2
x =[]
y= []
r = 4
center = [2,2]
theta_start = 0.0
theta_end = 2*math.pi
step_size= 0.125
steps = math.ceil((theta_end-theta_start)/step_size)
theta = theta_start
for i in range(0,steps):
  x.append(r*math.cos(theta)+center[0])
  y.append(r*math.sin(theta)+center[1])
  theta += step_size
plt.figure()
plt.plot(center[0],center[1],'.')
plt.plot(x,y,'-')
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
ax =plt.gca()
ax.set_aspect('equal',adjustable='box')
plt.show()
```

```python
# Question 3
import matplotlib.pyplot as plt
import math
x = []
y = []
a=3
b=5
center =[2,2]
theta_start = 0.0
theta_end = 3*math.pi
step_size = 0.125
steps = math.ceil((theta_end -theta_start) / step_size)
theta= theta_start
for i in range(steps):
  x.append(a*math.cos(theta)+center[0])
  y.append(b*math.sin(theta)+center[1])
  theta += step_size
plt.figure()
plt.plot(center[0],center[1],'.')
plt.plot(x,y,'-')
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ellipse centered at (2,2) with a=3 , b=5")
ax= plt.gca()
ax.set_aspect('equal',adjustable='box')
plt.show()
```

```python
# Question 4
import matplotlib.pyplot as plt
import math
q=0
p=0
a=0.5
step_size = 0.1
y_start = p-5
y_end = p + 5
x_values =[]
y_values =[]
y = y_start
while y <= y_end:
  x = 4*a*(y-p)**2+p
  x_values.append(x)
  y_values.append(y)
  y += step_size
plt.figure()
plt.title(f"Parabola with vertex at ({p},{q}) and a ={a}")
plt.plot(x_values,y_values,'-')
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Exercise 9
import matplotlib.pyplot as plt
import math
x = []
y = []
r = 4
a = 2
b = 2
theta = 0
while theta <= 2*math.pi:
    x.append(r * math.cos(theta) + a)
    y.append(r * math.sin(theta) + b)
    theta += 0.1
plt.plot(x, y)
plt.scatter(a, b)  # center
plt.gca().set_aspect('equal')
plt.grid()
plt.title("Circle")
plt.show()
```

```python
#Exercise 10
import matplotlib.pyplot as plt
import math
# Circle
xc, yc = [], []
theta = 0
while theta <= 2*math.pi:
    xc.append(4*math.cos(theta)+2)
    yc.append(4*math.sin(theta)+2)
    theta += 0.1
# Ellipse
xe, ye = [], []
theta = 0
while theta <= 2*math.pi:
    xe.append(3*math.cos(theta)+2)
    ye.append(5*math.sin(theta)+2)
    theta += 0.1
plt.plot(xc, yc, label="Circle")
plt.plot(xe, ye, label="Ellipse")
plt.legend()
plt.gca().set_aspect('equal')
plt.grid()
plt.show()
```

```python
#Exercise 11
import matplotlib.pyplot as plt
x_vals = []
y_vals = []
p = 0
q = 0
a = 0.5
y = -10
while y <= 10:
    x = (y - p)**2 / (4*a) + q
    x_vals.append(x)
    y_vals.append(y)
    y += 0.1
plt.plot(x_vals, y_vals)
plt.title("Parabola")
plt.grid()
plt.show()
```
