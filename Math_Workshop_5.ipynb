```python
#Task 6
import numpy as np
import matplotlib.pyplot as plt
U = np.array([-0.5, 0.5])
V = np.array([1,1])
W = U + V
def magnitude(v):
    return np.sqrt(np.sum(v**2))
def direction_cosines(v):
    mag = magnitude(v)
    return v / mag
def dot_product(a, b):
    return np.dot(a, b)
def cross_product(a, b):
    return np.cross(a, b)
print("U + V =", W)
print("Magnitude of U:", magnitude(U))
print("Direction cosines of U:", direction_cosines(U))
print("Dot product:", dot_product(U, V))
print("Cross product:", cross_product(U, V))
```
