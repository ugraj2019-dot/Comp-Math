```python
#Exercise 2
def is_prime(n):
  if n<2:
    return False
  for i in range(2, int(n**0.5)+1):
    if n%i==0:
      return False
  return True
print(is_prime(7))
```

```python
#Exercise 3
def twin_primes(limit):
  for i in range(3, limit, 2):
    if is_prime(i) and is_prime(i+2):
      print(i, i+2)
twin_primes(1000)
```

```python
#Exercise 4
def sumOfDivisors(n):
  total=0
  for i in range(1, n):
    if n%i==0:
      total+=i
  return total
print(sumOfDivisors(6))
```

```python
#Exercise 5
def next_perfect(start):
  n=start+1
  while True:
    if sumOfDivisors(n)==n:
      return n
    n+=1
print(next_perfect(6))
```

```python
#Exercise 6
marks=[20, 30, 35, 55, 70, 80, 90]
mean=sum(marks)/len(marks)

marks.sort()
n=len(marks)
if n%2==1:
  median=marks[n//2]
else:
  median=(marks[n//2-1]+marks[n//2])/2
print("Mean:", mean)
print("Median:", median)
```

```python
#Exercise 7
def decimal_to_binary(n):
  return bin(n)[2:]
print(decimal_to_binary(10))
```

```python
#Exercise 8
def prodDigits(n):
  product=1
  while n>0:
    product *= n%10
    n//=10
  return product
print(prodDigits(123))
```

```python
#Exercise 9
def multiply(*args):
  result=1
  for i in args:
    result *= i
  return result
print(multiply(1,2,3,4,5))
print(multiply(12,13,14))
```
