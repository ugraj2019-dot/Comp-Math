```python
#Task 2
import matplotlib.pyplot as plt
y1=[8,150,80]
y2=[54,77,54]
y3=[93,32,100]
y4=[116,11,76]
y5=[137,6,93]
y6=[184,1,72]
groups=['Bears', 'Dolphins', 'Whales']
data=[y1,y2,y3,y4,y5,y6]
years=[y1, y2, y3, y4, y5, y6]
for i in range(len(years)):
    plt.figure()
    plt.pie(data[i], labels=groups)
    plt.title("Animal Population")
    plt.show()
    plt.figure(figsize=(10,8))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.pie(data[i], labels=groups)
    plt.title(years[i])
plt.tight_layout()
plt.show()
```

```python
#Task 4(d)
import matplotlib.pyplot as plt
cars = ["Red", "Silver", "Blue", "White", "Green"]
numbers = [123, 230, 78, 193, 12]
colors = ["red", "silver", "blue", "white", "green"]
plt.bar(cars, numbers, color=colors)
plt.title("Car Colours Seen")
plt.xlabel("Car Colours")
plt.ylabel("Number Seen")
plt.show()
```

```python
import matplotlib.pyplot as plt
cars = ["Red", "Silver", "Blue", "White", "Green"]
numbers = [123, 230, 78, 193, 12]
plt.pie(numbers, labels=cars)
plt.title("Car Color Distribution")
plt.show()
```

```python
#Task 4(f) (A)
import matplotlib.pyplot as plt
cars = ["Red", "Silver", "Blue", "White", "Green"]
numbers = [123, 230, 78, 193, 12]
myexplode = [0.3, 0, 0, 0, 0]
plt.pie(numbers, labels=cars, explode=myexplode)
plt.show()
```

```python
#Task 4 (f) (B)
import matplotlib.pyplot as plt
cars = ["Red", "Silver", "Blue", "White", "Green"]
numbers = [123, 230, 78, 193, 12]
plt.pie(numbers, labels=cars, explode=myexplode, shadow=True)
plt.show()
```

```python
#Task 4(f) (C)
import matplotlib.pyplot as plt
cars = ["Red", "Silver", "Blue", "White", "Green"]
numbers = [123, 230, 78, 193, 12]
plt.pie(numbers, labels=cars)
plt.legend(title="Car Colours:")
plt.show()
```
