One parameter will create a sequence, one-by-one, from zero to one less than the parameter.
Two parameters will create a sequence, one-by-one, from the first parameter to one less than the second parameter.
Three parameters will create a sequence starting with the first parameter and stopping before the second parameter, but this time increasing each step by the third parameter.

```python
#Exercise 1
def fib(n):
  a = 0
  b = 1
  count = 0
  while count < n:
    print(a)
    c = a + b
    a = b
    b = c
    count += 1
n = int(input("Enter number of terms: "))
fib(n)
```

```python
#Exercise 2
def myPow(x, n):
  result = 1
  i = 0
  while i < n:
    result = result * x
    i += 1
  return result
x = int(input("Enter number of terms: "))
n = int(input("Enter number of terms: "))
print(myPow(x, n))
```

```python
#Exercise 3
def SumDigits(n):
  s = 0
  while n>0:
    digit = n%10
    s = s+digit
    n = n // 10
  print("Sum of digits:", s)
SumDigits(234)
```

```python
#Exercise 4
def squaredSum(arr):
  total = 0
  for num in arr:
    total=total+(num*num)
  return total
print(squaredSum([2,3,4]))
```

```python
#Exercise 5
def sumDigits(num):
  total=0
  while num>0:
    digit = num%10
    total=total+digit
    num=num//10
  return total
print(sumDigits(678))
```
